##### On their own these values don't give us much information. Maybe the basement rating and basement area can be combined into a single feature. Check for correlation

```python
basement_1_df = df[['BsmtFinSF1', 'BsmtFinType1', 'SalePrice']]
```

```python
basement_rating_dictionary={'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1, 'NA':0}
```

```python
basement_1_df['BsmtFinType1'].replace(basement_rating_dictionary, inplace=True)
```

```python
basement_1_df['Bsmt1'] = basement_1_df['BsmtFinType1'] * basement_1_df['BsmtFinSF1']
```

```python
basement_1_df.corr()
```

```python
sbn.lineplot(x=basement_1_df['Bsmt1'], y=df['SalePrice'])
```

##### It looks promising as an attempt to improve furthermore the model and reduce dimensionality
